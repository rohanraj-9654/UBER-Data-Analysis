### Variation of avg Trips/vehicles with dates and base number

```python
# Finding the ratio of trips/active_vehicles
uber_foil['trips/vehicle'] = uber_foil['trips']/uber_foil['active_vehicles']
```

```python
uber_foil.head()
```

```python
plt.figure(figsize=(16,8))
uber_foil.set_index('date').groupby(['dispatching_base_number'])['trips/vehicle'].plot()
plt.ylabel('Average trips/vehicle')
plt.title('Demand vs Supply chart (Date-wise)')
plt.legend()
```

#### From here we can analyse although base 'B02764' had more number of active vehicles and no. of trips than other bases, but the avg trips/vehicle is less than B02598, B02617, B02682.
#### Hence, these 3 bases having higher potential for business are given less attention currently, and must be given equal attention by increasing number of vehicles in those areas
